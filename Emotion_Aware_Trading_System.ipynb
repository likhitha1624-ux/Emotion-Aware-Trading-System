```python
sentiment_df = sentiment_df.dropna()
trader_df = trader_df.dropna()
```

```python
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])
trader_df['Timestamp IST'] = pd.to_datetime(trader_df['Timestamp IST'], format='%d-%m-%Y %H:%M')
```

```python
trader_df['Date'] = trader_df['Timestamp IST'].dt.date
sentiment_df['Date'] = sentiment_df['date'].dt.date
```

```python
sentiment_df.head()
trader_df.head()
```

```python
merged = pd.merge(trader_df, sentiment_df, on='Date')
```

```python
merged.head()
```

```python
merged['Sentiment_Score'] = merged['classification'].map({
    'Extreme Fear': -2,
    'Fear': -1,
    'Neutral': 0,
    'Greed': 1,
    'Extreme Greed': 2
})
```

```python
def strategy(row):
    if row['classification'] == 'Fear' and row['Closed PnL'] > 0:
        return "BUY"
    elif row['classification'] == 'Greed' and row['Closed PnL'] < 0:
        return "SELL"
    else:
        return "HOLD"

merged['Strategy'] = merged.apply(strategy, axis=1)
```

```python
merged[['Date','classification','Closed PnL','Strategy']].head()
```

```python
import matplotlib.pyplot as plt

merged.groupby('classification')['Closed PnL'].mean().plot(kind='bar')

plt.title("Profit based on Market Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Average Profit")
plt.show()
```

```python
merged.groupby('Strategy')['Closed PnL'].mean().plot(kind='bar')

plt.title("Strategy Performance")
plt.xlabel("Strategy")
plt.ylabel("Profit")
plt.show()
```

```python
merged.groupby('classification')['is_profit'].mean().plot(kind='bar')

plt.title("Win Rate by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Win Rate")
plt.show()
```

```python
merged['classification'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Trading Activity Distribution")
plt.show()
```

```python
import seaborn as sns

pivot = merged.pivot_table(
    values='Closed PnL',
    index='Side',
    columns='classification',
    aggfunc='mean'
)

sns.heatmap(pivot, annot=True)
```
